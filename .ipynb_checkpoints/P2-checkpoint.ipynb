```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import glob
%matplotlib inline
```

```python
class Line():
    def __init__(self):
        # was the line detected in the last iteration?
        self.detected = False    
        #polynomial coefficients averaged over the last n iterations
        self.best_fit = None  
        #polynomial coefficients for the most recent fit
        self.current_fit = [np.array([False])]  
        #radius of curvature of the line in some units
        self.radius_of_curvature = None 
        #distance in meters of vehicle center from the line
        self.line_base_pos = None 
        #x values for detected line pixels
        self.allx = [[]]
        #y values for detected line pixels
        self.ally = [[]]
```

```python
left_lane = Line()
right_lane = Line()
```

```python
def grayscale(image, isCV=True):
    ''' This method is used to convert a color image to 
        gray.
    '''
    if isCV:
        return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:
        return cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
```

### Camera calibration

```python
file_name_pattern = './camera_cal/calibration*.jpg'
    
images = glob.glob(file_name_pattern)

imgpoints = []
objpoints = []

nrows = 9
ncols = 6

# Generating 'objp' which will be the (x, y, z) coordinates of the chessboard corners.
# and creating a 9X6 mesh out fo  x and y coordinate assuming the z=0 
objp = np.zeros((9*6, 3), np.float32)
objp[:,:2] = np.mgrid[0:9, 0:6].T.reshape(-1, 2)

for fname in images:
    img = cv2.imread(fname)
    gray = grayscale(img)

    # Find the chessboard corners in distorted board images
    retVal, corners = cv2.findChessboardCorners(gray, (nrows, ncols), None)

    if retVal:
        imgpoints.append(corners)
        objpoints.append(objp)
        
# Calibrating camera using the calculated image points i.e. corners of the chessboard images
retVal, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, gray.shape[::-1], None, None)
```

### Distortion correction

```python
def undistort_image(image, calibration_matrix, distortion_coeff):
    ''' This method undistort the provided image using calibration matrix (3x3) and 
        distortion coefficients
        Return: Undistorted Image
    '''
    return cv2.undistort(image, calibration_matrix, distortion_coeff, None, calibration_matrix)
```

### Color/gradient threshold

```python
def sobel_x_threshold(image, kernel=3, thresh=(0,255)):
    ''' This method calculates the image gradient in the x direction
        with the provided kernel size, filter out the required pixel
        range with thresh parameter and returns back a binary thresholded
        image.
    '''
    gray = grayscale(image, False)
    sobel = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=kernel)
    sobel_abs = np.absolute(sobel)
    sobel_ = np.uint8(255*sobel_abs/np.max(sobel_abs))
    sobel_binary = np.zeros_like(sobel_)
    sobel_binary[(sobel_ >= thresh[0]) & (sobel_ < thresh[1])] = 1
    return sobel_binary
```

```python
def sobel_mag_thresh(image, kernel=3, thresh=(0, 255), debug=False):
    ''' This method make binary image of gradient based on magnitude thresholds.
    '''
    gray = grayscale(image, False)
    sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=kernel)
    sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=kernel)
    mag = np.sqrt(sobelx**2 + sobely**2)
    mag = (mag/np.max(mag)*255).astype(np.uint8) 
    binary_output = np.zeros_like(mag)
    binary_output[(mag >= thresh[0]) & (mag <= thresh[1])] = 1
    return binary_output
```

```python
def hls_threshold(image, thresh=(0, 255)):
    ''' This method convert the image color space from RGB to HLS 
        and extracts the image S channel. it filters out the required 
        pixel range with thresh parameter and returns back a binary 
        thresholded image.
    '''
    hls = cv2.cvtColor(image, cv2.COLOR_RGB2HLS)
    s_channel = hls[:,:,2]
    binary_s = np.zeros_like(s_channel)
    
    binary_s[(s_channel > thresh[0]) & (s_channel <= thresh[1])] = 1
    return binary_s
```

### Perspective transform

```python
def warp_image(image):
    ''' The method will perform the perspective transformation of the provided image
        by calculating the 3x3 perspective transform matrix. 
        It also calculates the inverse of the perspective transform matrix to get the 
        original results back.
        Return: Warped image, Inverse perspective transform matrix
    '''
    offset = 150
    img_size = (image.shape[1], image.shape[0])
    
    src_points = np.array([[730, 470], [1130, 690], [230, 690], [560, 470]], np.float32)
    dst_points = np.array([[img_size[0] - offset, 0], [img_size[0] - offset, img_size[1]], 
                           [offset, img_size[1]], [offset, 0]], np.float32)

    warp_M = cv2.getPerspectiveTransform(src_points, dst_points)
    Minv = cv2.getPerspectiveTransform(dst_points, src_points)
    
    output_image = cv2.warpPerspective(image, warp_M, img_size, flags=cv2.INTER_LINEAR)
    return output_image, Minv
```

### Detect lane lines

```python
def find_lane_pixels(image):
    ''' This method detects the left and right lane pixels 
        and returns the x and y coordinates for both left and 
        right lanes.
    '''
    
    # Calculate the histogram of the bottom half of image by summing all
    # values in that part of image
    hist = np.sum(image[image.shape[0]//2:,:], axis=0)
    
    # Created colored image of the given binary 'image' for plotting purposes
    out = np.dstack((image, image, image))*255
    
    # Calculates the midpoint of the road to separates left and right lanes
    midpoint = np.int(hist.shape[0]//2)
    left_base = np.argmax(hist[:midpoint])
    right_base = np.argmax(hist[midpoint:]) + midpoint
    
    # Total no. of windows to draw on the lanes
    nwindow = 9
    # Left and right side margin of the window
    margin = 100
    # Minimum no. of points to recalculate the center
    minpix = 50
    
    # Getting the height of each window wrt the image height
    window_height = np.int(image.shape[0]//nwindow)
    
    # Retrieve all the non zero coordinates on the binary image
    nonzero = image.nonzero()
    # Separates x and y non zero coordinate points
    nonzeroy = np.array(nonzero[0])
    nonzerox = np.array(nonzero[1])
    
    # Initializing the currnet pointers with left and right lane values
    left_current = left_base
    right_current = right_base
    
    # To store lanes pixel indices
    left_curve_inds = []
    right_curve_inds = []
    
    for window in range(nwindow):
        # Calculating low and high y point for current window
        win_y_low = image.shape[0] - (window+1)*window_height
        win_y_high = image.shape[0] - window*window_height
        
        # Calculating low and high x point for left lane current window
        win_leftx_low = left_current - margin
        win_leftx_high = left_current + margin
        
        # Calculating low and high x point for right lane current window
        win_rightx_low = right_current - margin
        win_rightx_high = right_current + margin

        # Drawing the rectangles to display left and right lane windows
        cv2.rectangle(out, (win_leftx_low, win_y_low), (win_leftx_high, win_y_high), (0, 0, 255), 2)
        cv2.rectangle(out, (win_rightx_low, win_y_low), (win_rightx_high, win_y_high), (0, 0, 255), 2)
        
        # Indices of ON(1) pixels which falls under the left lane current window limits
        left_range_px = ((nonzerox >= win_leftx_low) & (nonzerox <= win_leftx_high) &
                         (nonzeroy >= win_y_low) & (nonzeroy <= win_y_high)).nonzero()[0]
        
        # Indices of ON(1) pixels which falls under the right lane current window limits
        right_range_px = ((nonzerox >= win_rightx_low) & (nonzerox <= win_rightx_high) &
                         (nonzeroy >= win_y_low) & (nonzeroy <= win_y_high)).nonzero()[0]

        # Storing selected pixels
        left_curve_inds.append(left_range_px)
        right_curve_inds.append(right_range_px)
        
        # If the total no. of ON(1) pixels is greater than the minpix value then calculating
        # the mean of their indices to recenter the window (both left and right lane)
        if len(left_range_px) > minpix:
            left_current = np.int(np.mean(nonzerox[left_range_px]))
        if len(right_range_px) > minpix:
            right_current = np.int(np.mean(nonzerox[right_range_px]))
            
    try:
        # Transforming all captured indices into one single dimension array
        left_curve_inds = np.concatenate(left_curve_inds)
        right_curve_inds = np.concatenate(right_curve_inds)
    except:
        pass
    
    # Retriving the left and right lane `x` and `y` pixel indices
    leftx = nonzerox[left_curve_inds]
    lefty = nonzeroy[left_curve_inds]
    rightx = nonzerox[right_curve_inds]
    righty = nonzeroy[right_curve_inds]
    
    # Fitting 2nd order polynomial
    if len(leftx) == len(lefty):
        left_lane.current_fit = np.polyfit(lefty, leftx, 2)
    if len(rightx) == len(righty):
        right_lane.current_fit = np.polyfit(righty, rightx, 2)   
    
    return leftx, lefty, rightx, righty, out
```

```python
def find_existing_pixels(image):
    ''' This method will search for lane line pixels on the basis of
        polynomial coefficients already provided
    '''
    margin = 100
    
    left_fit = left_lane.current_fit
    right_fit = right_lane.current_fit
    
    out = np.dstack((image, image, image))*255
    
    # Retrieve all the non zero coordinates on the binary image
    nonzero = image.nonzero()
    # Separates x and y non zero coordinate points
    nonzeroy = np.array(nonzero[0])
    nonzerox = np.array(nonzero[1])
    
    # Indices of ON(1) pixels which falls under the left lane current window limits
    left_lane_inds = ((nonzerox >= (left_fit[0]*nonzeroy**2 + left_fit[1]*nonzeroy + left_fit[2] - margin)) & 
                     (nonzerox <= (left_fit[0]*nonzeroy**2 + left_fit[1]*nonzeroy + left_fit[2] + margin)))

    # Indices of ON(1) pixels which falls under the right lane current window limits
    right_lane_inds = ((nonzerox >= (right_fit[0]*nonzeroy**2 + right_fit[1]*nonzeroy + right_fit[2] - margin)) & 
                      (nonzerox <= (right_fit[0]*nonzeroy**2 + right_fit[1]*nonzeroy + right_fit[2] + margin)))
    
    # Retriving the left and right lane `x` and `y` pixel indices
    leftx = nonzerox[left_lane_inds]
    lefty = nonzeroy[left_lane_inds] 
    rightx = nonzerox[right_lane_inds]
    righty = nonzeroy[right_lane_inds]
    
    # Fitting 2nd order polynomial
    if len(leftx) == len(lefty):
        left_lane.current_fit = np.polyfit(lefty, leftx, 2)
    if len(rightx) == len(righty):
        right_lane.current_fit = np.polyfit(righty, rightx, 2)   
        
    return leftx, lefty, rightx, righty
```

```python
def check_line_validity(image):
    ''' This method checks whether the lines generated are parallel or not.
        If the lines are not parallel for more than the allowed limit then
        it will set line detected to False and sliding window algorithm will 
        generate new lines from scratch.
    '''
    maximum_error_allowed = 150 #px
    
    left_fit = left_lane.current_fit
    right_fit = right_lane.current_fit
    
    ploty = np.linspace(0, image.shape[0]-1, image.shape[0])
    
    try:
        left_lane_x = left_fit[0]*ploty**2 + left_fit[1]*ploty + left_fit[2] 
        right_lane_x = right_fit[0]*ploty**2 + right_fit[1]*ploty + right_fit[2]
    except:
        pass
    
    if len(left_lane_x) == len(right_lane_x):
        minimum_distance = np.min(np.absolute(left_lane_x - right_lane_x))
        maximum_distance = np.max(np.absolute(left_lane_x - right_lane_x))
    
        if maximum_distance - minimum_distance < maximum_error_allowed:
            left_lane.detected = True
            right_lane.detected = True
        else:
            left_lane.detected = False
            right_lane.detected = False
```

```python
def perform_line_smoothening(leftx, lefty, rightx, righty):
    ''' This method will store few previous and current frames lane line pixels
        and create 2D polynomial coefficients to best fit the lane lines.
    '''
    max_points = 200
    if left_lane.detected and right_lane.detected:
        left_lane.allx.append(leftx)
        left_lane.ally.append(lefty)
        right_lane.allx.append(rightx)
        right_lane.ally.append(righty)
        
        if len(left_lane.allx) > max_points:
            left_lane.allx = left_lane.allx[len(left_lane.allx)-max_points:]
            left_lane.ally = left_lane.ally[len(left_lane.ally)-max_points:]
            
        if len(right_lane.allx) > max_points:
            right_lane.allx = right_lane.allx[len(right_lane.allx)-max_points:]
            right_lane.ally = right_lane.ally[len(right_lane.ally)-max_points:]    
        
        left_all_x_pts = np.concatenate(left_lane.allx)
        left_all_y_pts = np.concatenate(left_lane.ally)
        right_all_x_pts = np.concatenate(right_lane.allx)
        right_all_y_pts = np.concatenate(right_lane.ally)
        
        if len(left_all_x_pts) == len(left_all_y_pts):
            left_lane.best_fit = np.polyfit(left_all_y_pts, left_all_x_pts, 2)
        if len(right_all_x_pts) == len(right_all_y_pts):
            right_lane.best_fit = np.polyfit(right_all_y_pts, right_all_x_pts, 2)
    else:
        left_lane.best_fit = left_lane.current_fit
        right_lane.best_fit = right_lane.current_fit
```

### Determine the lane curvature

```python
def calculate_curvature_and_position(leftx, lefty, rightx, righty, image):
    ''' This method will calculate the radius of curvature of the road 
        In real world measurments
    '''
    lane_length_mtrs = 30
    lane_width_mtrs = 3.7
    
    # 720 as the height of images is 720 px. Returns meter/px in y dim.
    y_mtrs_per_px = lane_length_mtrs/720 
    # 980 as the width of lane lines in warped image is 980 px (1130-150).Returns meter/px in x dim.
    x_mtrs_per_px = lane_width_mtrs/980
    
    # Fit a second order polynomial to each lane line
    left_fit = np.polyfit(lefty*y_mtrs_per_px, leftx*x_mtrs_per_px, 2)
    right_fit = np.polyfit(righty*y_mtrs_per_px, rightx*x_mtrs_per_px, 2)
    
    # Getting maximum y-value to calculate the radius of curvature at bottom of the image
    #y_eval = np.max(ploty)
    y_eval = image.shape[0]
    #print(left_fit, right_fit)
    # Calculating radius of curvature
    left_lane.radius_of_curvature = ((1 + (2*left_fit[0]*y_eval*y_mtrs_per_px + 
                                           left_fit[1])**2)**3/2) / np.absolute(2*left_fit[0])
    right_lane.radius_of_curvature = ((1 + (2*right_fit[0]*y_eval*y_mtrs_per_px + 
                                            right_fit[1])**2)**3/2) / np.absolute(2*right_fit[0])
    #print(left_lane.radius_of_curvature, right_lane.radius_of_curvature)
    # To calculate position of vehicle, we can take height of the image and calculate `x` 
    # position by fitting 2nd degree polynomial curve equation with `y`
    left_x = left_fit[0]*y_eval*y_mtrs_per_px**2 + left_fit[1]*y_eval*y_mtrs_per_px + left_fit[2]
    right_x = right_fit[0]*y_eval*y_mtrs_per_px**2 + right_fit[1]*y_eval*y_mtrs_per_px + right_fit[2]
    
    # Image center will be vehicle center
    image_center = (image.shape[1]/2) * x_mtrs_per_px
    
    left_lane.line_base_pos = image_center - left_x
    right_lane.line_base_pos = right_x - image_center
    
    if right_lane.line_base_pos < left_lane.line_base_pos:
        side = "right"
    elif right_lane.line_base_pos > left_lane.line_base_pos:
        side = "left"
    else:
        side = "center"
        
    return side
```

```python
def draw_mapping(warped, Minv, undist, side):
    '''Create an image to draw the lines on
    '''
    # Fitting 2nd order polynomial
    left_fit = left_lane.best_fit
    right_fit = right_lane.best_fit
    
    #left_fit = left_lane.current_fit
    #right_fit = right_lane.current_fit
    
    # Generating `y` points wrt the height of the image
    ploty = np.linspace(0, warped.shape[0]-1, warped.shape[0])
    
    try:
        # Calculating left and right `x` coordinate values as a function of 
        # 2nd order `y` polynomial
        leftx = left_fit[0]*ploty**2 + left_fit[1]*ploty + left_fit[2]
        rightx = right_fit[0]*ploty**2 + right_fit[1]*ploty + right_fit[2]
    except:
        pass
    
    warp_zero = np.zeros_like(warped).astype(np.uint8)
    color_warp = np.dstack((warp_zero, warp_zero, warp_zero))

    # Recast the x and y points into usable format for cv2.fillPoly()
    pts_left = np.array([np.transpose(np.vstack([leftx, ploty]))])
    pts_right = np.array([np.flipud(np.transpose(np.vstack([rightx, ploty])))])
    pts = np.hstack((pts_left, pts_right))

    # Draw the lane onto the warped blank image
    cv2.fillPoly(color_warp, np.int_([pts]), (0,255, 0))

    # Warp the blank back to original image space using inverse perspective matrix (Minv)
    newwarp = cv2.warpPerspective(color_warp, Minv, (warped.shape[1], warped.shape[0])) 
    # Combine the result with the original image
    result = cv2.addWeighted(undist, 1, newwarp, 0.3, 0)
    
    vehicle_pos = np.round(np.absolute(left_lane.line_base_pos - right_lane.line_base_pos), 2)
    
    cv2.putText(result, f'Radius of Curvature={(left_lane.radius_of_curvature+right_lane.radius_of_curvature)/2:.1f}(m)',
                  (20, 70), cv2.FONT_HERSHEY_SIMPLEX, 2.5, (255,255,255), 4, cv2.LINE_AA)
    #cv2.putText(result, f'Radius of Curvature={right_lane.radius_of_curvature:.1f}(m)',
    #              (20, 210), cv2.FONT_HERSHEY_SIMPLEX, 2.5, (255,255,255), 4, cv2.LINE_AA)
    cv2.putText(result, f'Vehicle is {vehicle_pos:.2f}m {side} of Center',
                  (20, 140), cv2.FONT_HERSHEY_SIMPLEX, 2.5, (255,255,255), 4, cv2.LINE_AA)
    return result
```

```python
def pipeline(image):
    undist_image = undistort_image(image, mtx, dist)
    x = sobel_x_threshold(image, thresh=(50, 150))
    s = hls_threshold(image, thresh=(150, 255))
    mag = sobel_mag_thresh(undist_image, kernel=11, thresh=(120, 255))
    combined = np.zeros_like(x)
    combined[(x == 1) | (s == 1) | (mag == 1)] = 1
    warped_image, Minv = warp_image(combined)
    if left_lane.detected or right_lane.detected: 
        leftx, lefty, rightx, righty = find_existing_pixels(warped_image)
    else:
        leftx, lefty, rightx, righty, _ = find_lane_pixels(warped_image)
    check_line_validity(warped_image)
    perform_line_smoothening(leftx, lefty, rightx, righty)
    side = calculate_curvature_and_position(leftx, lefty, rightx, righty, warped_image)
    return draw_mapping(warped_image, Minv, undist_image, side)
```
